### Modelo sugere uma carteira que atende as expectativas do usuário

```python
import numpy as np
import pandas as pd
import pandas_datareader as web
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

start = datetime(2019, 1, 1)

symbols = ['^BVSP','PETR4.SA', 'PTBL3.SA', 'VALE3.SA']

df = pd.DataFrame()

for ticker in symbols: 
    r = web.DataReader(ticker, "yahoo", start)[['Close']].pct_change()
    df[ticker + 'Return'] = r['Close']

df = df.dropna()
df.head()
```

```python
sns.distplot(df['^BVSPReturn'])
```

```python
sns.set()
df['^BVSPReturn'].plot()
```

```python
((df / df.iloc[0]) * 100).plot()
```
